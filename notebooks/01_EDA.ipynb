```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/sales_data_sample.csv", encoding='latin1')
```

# ðŸ§ª Ã‰TAPE 2 â€“ Nettoyage & prÃ©paration de base

```python
# Conversion de ORDERDATE
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Extraire mois, annÃ©e, trimestre
df['YEAR'] = df['ORDERDATE'].dt.year
df['MONTH'] = df['ORDERDATE'].dt.month
df['QUARTER'] = df['ORDERDATE'].dt.quarter
```

##### ðŸ§¼ VÃ©rifier les valeurs manquantes

```python
df.isna().sum()
```

##### ðŸ“Š Visualiser les ventes par mois

```python
monthly_sales = df.groupby(df['ORDERDATE'].dt.to_period('M'))['SALES'].sum().reset_index()
monthly_sales['ORDERDATE'] = monthly_sales['ORDERDATE'].astype(str)
monthly_sales['ORDERDATE'] = pd.to_datetime(monthly_sales['ORDERDATE'])

plt.figure(figsize=(10, 4))
sns.lineplot(data=monthly_sales, x='ORDERDATE', y='SALES', marker='o')
plt.title("Ã‰volution mensuelle des ventes")
plt.xlabel("Date")
plt.ylabel("Total des ventes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### ðŸ“Š Visualiser les ventes par produit ou pays

```python
# Par ligne de produit
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='PRODUCTLINE', y='SALES', estimator=sum)
plt.title("Ventes totales par ligne de produit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Par pays
plt.figure(figsize=(10, 4))
top_countries = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title("Top 10 des pays par ventes totales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
